```python
from sklearn.datasets import make_circles


nSamples = 1000

x, y = make_circles(nSamples, noise=0.03, random_state = 42)


#We have two inputs and one output
x, y
```

```python
import pandas as pd
import tensorflow as tf
model = tf.keras.Sequential([
        tf.keras.layers.Dense(1)
])



model.compile(loss = tf.keras.losses.BinaryCrossentropy(),
              optimizer = tf.keras.optimizers.Adam(lr=0.1),
              metrics=['accuracy'])

model.fit(x, y, epochs=10)

model.evaluate(x, y)
```

```python
import numpy as np

tf.random.set_seed(42)
#creating a mesh grid
X = tf.range(1,1000,5)
Y = tf.range(100, 1100,5)

X = tf.cast(X, dtype = tf.float32)
Y = tf.cast(Y, dtype = tf.float32)

xTrain = X[:150]
xTest = X[150:]

yTrain = Y[:150]
yTest = Y[150:]


model = tf.keras.Sequential([
        tf.keras.layers.Dense(1, input_shape = [1])
])

model.compile(loss=tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(lr =0.1),
              metrics = ['mae'])


xTrain, yTrain
model.fit(xTrain, yTrain, epochs=100)
```

```python
import matplotlib.pyplot as plt
pred = model.predict(xTest)

plt.figure(figsize =(8,7))

plt.scatter(xTrain, yTrain, c="b", label="Training")
plt.scatter(xTest, yTest, c="y", label="Testing")
plt.scatter(xTest, pred, c = "r",label = "predictions")

plt.legend()

pred
```
