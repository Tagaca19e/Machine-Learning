```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles


X = tf.range(0, 1000, 5)
Y = tf.range(100, 1100, 5)


xTrain = X[:150]
xTest = X[150:]

yTrain = Y[:150]
yTest = Y[150:]

nCircles = 1000

x, y = make_circles(nCircles, noise = 0.03, random_state = 42)



tf.random.set_seed(42)


model = tf.keras.Sequential([
        tf.keras.layers.Dense(1, activation="linear")        
])

model.compile(loss="binary_crossentropy",
              optimizer = tf.keras.optimizers.Adam(lr=0.1),
              metrics=['accuracy'])

model.fit(x, y, epochs=100, verbose =1)
```

```python
tf.random.set_seed(42)


model_2 = tf.keras.Sequential([
          tf.keras.layers.Dense(1, activation = "relu")
])


model_2.compile(loss = tf.keras.losses.BinaryCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(lr=0.1),
                metrics=['accuracy'])

model_2.fit(x, y, epochs=100)
```

```python
model_3 = tf.keras.Sequential([
                               
          tf.keras.layers.Dense(4, activation="relu"),
          tf.keras.layers.Dense(4, activation="relu"),
          tf.keras.layers.Dense(1, activation = "sigmoid")
])

model_3.compile(loss="binary_crossentropy",
                optimizer=tf.keras.optimizers.Adam(lr =0.1),
                metrics=['accuracy'])

model_3.fit(x, y, epochs =100)

model_3.evaluate(x, y)
```

```python
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.2)


xTrain.shape, yTrain.shape, xTest.shape, yTest.shape

```

```python


nCircles = 1000

a, b = make_circles(nCircles, noise = 0.03, random_state = 42)


model_4 = tf.keras.Sequential([
          tf.keras.layers.Dense(4, activation="relu"),
          tf.keras.layers.Dense(4, activation="relu"),
          tf.keras.layers.Dense(1, activation="sigmoid")
])

model_4.compile(loss= tf.keras.losses.binary_crossentropy,
                optimizer = tf.keras.optimizers.Adam(lr = 0.1),
                metrics=['accuracy'])


model_4.fit(a, b, epochs=100)
```

```python
# Set random seed
tf.random.set_seed(42)

# Create a model
model_7 = tf.keras.Sequential([
  tf.keras.layers.Dense(4, activation= tf.keras.activations.relu), # hidden layer 1, ReLU activation
  tf.keras.layers.Dense(4, activation= tf.keras.activations.relu), # hidden layer 2, ReLU activation
  tf.keras.layers.Dense(1, activation= tf.keras.activations.sigmoid) # ouput layer, sigmoid activation
])

# Compile the model
model_7.compile(loss=tf.keras.losses.binary_crossentropy,
                optimizer=tf.keras.optimizers.Adam(lr = 0.01),
                metrics=['accuracy'])

# Fit the model
history = model_7.fit(xTrain, yTrain, epochs=100)

# model_7.evaluate(xTrain, yTrain)
```

```python
model_7.evaluate(xTrain, yTrain)


pd.DataFrame(history.history).plot()

pd.DataFrame(history.history)
```

```python
#Shows us the our learning rate

scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10 ** (epoch/20))
model_7.compile(loss="binary_crossentropy",
                optimizer = tf.keras.optimizers.Adam(lr = 0.001),
                metrics =['accuracy'])

hist = model_7.fit(xTrain, yTrain, epochs=100, callbacks=[scheduler])


pd.DataFrame(hist.history).plot()
```

```python
lrs = 1e-4 * (10 ** (tf.range(100)/20))
plt.figure(figsize=(10, 7))
plt.semilogx(lrs, hist.history["loss"])
plt.xlabel("learning rate")
plt.ylabel("loss")

```
